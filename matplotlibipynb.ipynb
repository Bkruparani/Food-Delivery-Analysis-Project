```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df = pd.read_excel('/content/Excel Food Delivery Project.xlsx')
```

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Test Plot")
plt.show()
```

```python
df['City'].value_counts().plot(kind='bar')
plt.title("Orders per City")
plt.show()
```

```python
df['Delivery_status'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Order Status")
plt.show()
```

```python
plt.hist(df['Rating'])
plt.title("Customer Ratings")
plt.show()
```

```python
plt.hist(df['Price'])
plt.title("Order Price Distribution")
plt.show()
```

```python
plt.hist(df['Quantity'])
plt.title("Quantity Ordered")
plt.show()
```

```python
plt.scatter(df['Price'], df['Rating'])
plt.xlabel("Price")
plt.ylabel("Rating")
plt.show()
```

```python
plt.scatter(df['Quantity'], df['Price'])
plt.show()
```

```python
df['Order_date'] = pd.to_datetime(df['Order_date'])
df.groupby(df['Order_date'].dt.day).size().plot()
plt.title("Orders per Day")
plt.show()
```

```python
df.groupby(df['Order_date'].dt.month).size().plot()
plt.title("Orders per Month")
plt.show()
```

```python
df['weekday'] = df['Order_date'].dt.day_name()
df['weekday'].value_counts().plot(kind='bar')
plt.show()
```

```python
df.groupby('City')['Price'].mean().plot(kind='bar')
plt.title("Avg Price per City")
plt.show()
```

```python
df.groupby('City')['Rating'].mean().plot(kind='bar')
plt.show()
```

```python
df['City'].value_counts().head(10).plot(kind='bar')
plt.show()
```

```python
plt.boxplot(df['Price'])
plt.title("Price Outliers")
plt.show()
```

```python
plt.boxplot(df['Rating'])
plt.show()
```

```python
plt.boxplot(df['Quantity'])
plt.show()
```

```python
df['Orders'] = 1
df['Orders'].cumsum().plot()
plt.title("Cumulative Orders")
plt.show()
```

```python
df_sorted = df.sort_values('Price')
plt.plot(df_sorted['Price'])
plt.show()
```
