```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Excel Food Delivery Project (1).csv')
df.head()
```

```python
sns.countplot(x='Delivery_status', data=df)
plt.title("Count of Delivery Status")
plt.show()
```

```python
pd.crosstab(df['Delivery_status'], df['Payment_method']).plot(kind='bar')

plt.title("Delivery Status vs Payment Method")
plt.xlabel("Delivery Status")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
sns.stripplot(x='Delivery_status', y='Order_frequency', data=df, size=4, jitter=True)

plt.title("Delivery Status vs Delivery Time")
plt.show()
```

```python
pd.crosstab(df['City'], df['Delivery_status']).plot(kind='bar')

plt.title("City vs Delivery Status")
plt.xlabel("City")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
