```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Excel Food Delivery Project (1).csv")
```

```python
df
```

```python
counts = df['Delivery_status'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(counts.index, counts.values)
plt.title("Delivery Status Count")
plt.xlabel("Status")
plt.ylabel("Number of Orders")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['Price'], bins=20, edgecolor='black')
plt.title("Delivery Time Distribution")
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Delivery_status'], df['Rating'], alpha=0.5)
plt.title("Delivery Time vs Rating")
plt.xlabel("Delivery Time")
plt.ylabel("Rating")
plt.show()
```

```python
city_counts = df['City'].value_counts()

plt.figure(figsize=(8,4))
plt.bar(city_counts.index, city_counts.values)
plt.xticks(rotation=45)
plt.title("Orders per City")
plt.xlabel("City")
plt.ylabel("Orders")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(df['Rating'])
plt.title("Rating Distribution")
plt.ylabel("Rating")
plt.show()
```

```python
status_counts = df['Delivery_status'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(status_counts.values, labels=status_counts.index, autopct='%1.1f%%')
plt.title("Order Status Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Price'], df['Order_frequency'], alpha=0.5)
plt.title("Price vs Order Frequency")
plt.xlabel("Price")
plt.ylabel("Order Frequency")
plt.show()
```

```python
df.groupby('Order_date').size().plot()
plt.show()
```

```python
df['Order_date'] = pd.to_datetime(df['Order_date'], format='mixed', errors='coerce')
df['weekday'] = df['Order_date'].dt.day_name()
df['weekday'].value_counts().plot(kind='bar')
plt.show()
```

```python
df.groupby(df['Order_date'].dt.month).size().plot()
plt.title("Orders per Month")
plt.show()
```

```python
plt.hist(df['Quantity'])
plt.title("Quantity Ordered")
plt.show()
```
